```python
#dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ML Projects/Project-1-Sonar-Mine-vs-Rock/sonar data.csv', header=None)
```

```python
df
```

```python
label=LabelEncoder()
```

```python
label.fit(df[60])
```

```python
df['target']=label.transform(df[60])
```

```python
X=df.drop(columns='target', axis=1)
```

```python
Y=df[60]
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
trained_model=model.predict(X_train)
acc=accuracy_score(trained_model, Y_train)
```

```python
acc
```

```python
test_model=model.predict(X_test)
acc=accuracy_score(test_model, Y_test)
```
