```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("/Users/srinivasdeekonda/Downloads/loan_dataset.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df=df.replace({'Loan_Status':{'Y':0, 'N':1}, 'Gender':{'Male':0, 'Female':1}, 'Marreid':{'Yes':0, 'No':1}, 'Dependents':{'3+':4}, 'Education':{'Graduate':0, 'Not Graduate':1}, 'Self_Employed':{'Yes':0, 'No':1}, 'Propert_Area':{'Rural':0, 'Urban':1, 'Semiurban':2}})
```

```python
df=df.replace({'Married':{'Yes':0, 'No':1},'Property_Area':{'Rural':0, 'Urban':1, 'Semiurban':2}})
```

```python
df
```

```python
df=df.drop(columns='Loan_ID')
```

```python
X=df.drop(columns='Loan_Status')
```

```python
Y=df['Loan_Status']
```

```python
X
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,random_state=3, test_size=0.2, stratify=Y)
```

```python
classifier=svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

```python
train_prediction=classifier.predict(X_train)
```

```python
train_test_accuracy= accuracy_score(Y_train, train_prediction)
```

```python
train_test_accuracy
```

```python
test_prediction=classifier.predict(X_test)
X_test_accuracy= accuracy_score(Y_test, test_prediction)
```

```python
X_test_accuracy
```
